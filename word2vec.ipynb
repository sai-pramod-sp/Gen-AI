# Custom Training for Word2Vec (Static wordEmbeddings)

```python
import pandas as pd
import numpy as np
import os
import gensim
import nltk
```

```python
nltk.download("all")
```

```python
from nltk import sent_tokenize                # To convert data to sentence
from gensim.utils import simple_preprocess    # Method which is responsible to preprocess the data
```

```python
story = []
```

```python
for file in os.listdir("/content/data"):
  print(file)
```

```python
fi = open("/content/data/001ssb.txt", encoding = 'unicode_escape')  # for opening the file
corpus = fi.read()                                                  # for reading the file
print(corpus)
```

```python
raw_sentence = sent_tokenize(corpus)               # Converting the raw_data to the sentence tokenization
```

```python
# code to do preprocessing and converting the sentence tokenization to the word_tokenization
for i in raw_sentence:
  print(simple_preprocess(i))
```

```python
for filename in os.listdir("/content/data"):
  file_path = os.path.join("/content/data", filename)              # To join the file path with the data folder path
  with open(file_path, encoding = "unicode_escape") as f:          # To open the file
    corpus = f.read()                                              # To read the file
  raw_sentence = sent_tokenize(corpus)                             # To convert the raw_data to the sentence_tokenization
  for sentence in raw_sentence:
    story.append(simple_preprocess(sentence))                      # To append all the corpus to the story which is already word tokenized by using simple_preprocess
```

```python
len(story)
```

```python
# window (int, optional) – Maximum distance between the current and predicted word within a sentence.
# vector_size (int, optional) – Dimensionality of the word vectors.
# min_count (int, optional) – Ignores all words with total frequency lower than this.
# Refer (https://radimrehurek.com/gensim/models/word2vec.html) for all the parameters from gensim

model = gensim.models.Word2Vec(window=10, min_count=5, vector_size=150)
```

```python
''' build_vocab ---> is a method used to construct the vocabulary for a Word2Vec model or other similar models.
    This method analyzes the input corpus and builds a vocabulary consisting of unique words along with their frequencies. The vocabulary is essential for
    training word embedding models, as it determines which words will be included in the model's dictionary and thus be represented as vectors. '''

model.build_vocab(story)
model.train(story, total_examples=model.corpus_count, epochs = 50)
```

```python
# like wise we can get the vector representation of any word
model.wv['king']             # Vector representation of king
len(model.wv['king'])        # length of the king
```

```python
model.wv.most_similar("filling")
```

```python
model.wv.doesnt_match(["jon", 'rikon', 'robb', 'arya', 'sansa', 'bran'])    # will give the result which is unique word in the present list
```

```python
vectors = model.wv.get_normed_vectors()          # to get the all the vectors
```

```python
vectors
```

```python
y = model.wv.index_to_key                             # returns all the words which are selected by the model
len(model.wv.index_to_key)                        # returns the length of the words
```
